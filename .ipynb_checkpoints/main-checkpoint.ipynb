## data preparation

```python
class power_law:
    def __init__(self, a, b, g):
        self.a = a
        self.b = b
        self.g = g
        
    def rvs(self, size=1, type='int'):
        """Power-law gen for pdf(x)\propto x^{g-1} for a<=x<=b"""
        r = np.random.random(size=size)
        ag, bg = self.a**self.g, self.b**self.g
        if type=='int':
            return [int(i) for i in (ag + (bg - ag)*r)**(1./self.g)]
        else:
            return np.round((ag + (bg - ag)*r)**(1./self.g), 2).tolist()

    def pdf(self, x):
        ag, bg = self.a**self.g, self.b**self.g
        return self.g * x**(self.g-1) / (bg - ag)

power = power_law(0, 131, 0.8)
sample = power.rvs(size=10000, type='float')
# sample
# sample = np.abs(np.random.normal(loc=10.0, scale=50, size=1000))
_, _, _ = plt.hist(sample, bins=100)
# plt.plot(np.arange(1, 118), power.pdf(np.arange(1, 118)))
```

```python
# generate training dataset
sample_size = 1000000
# probabilities from beta distribution
beta_rand_a = np.round(np.random.beta(0.5, 0.5, size=sample_size), 2)
beta_rand_b = np.round(np.random.beta(0.5, 0.5, size=sample_size), 2)
# values from power law distribution
power = power_law(0, 131, 0.8)
powerlaw_rand_a_int = power.rvs(size=int(sample_size*0.9), type='int') + [int(n) for n in np.random.uniform(-20, 0, size=int(sample_size*0.1))]
powerlaw_rand_b_int = power.rvs(size=int(sample_size*0.9), type='int') + [int(n) for n in np.random.uniform(-20, 0, size=int(sample_size*0.1))]
random.shuffle(powerlaw_rand_a_int)
random.shuffle(powerlaw_rand_b_int)
powerlaw_rand_a_flt = power.rvs(size=int(sample_size*0.9), type='float') + np.round(np.random.uniform(-20, 0, size=int(sample_size*0.1)), 2).tolist()
powerlaw_rand_b_flt = power.rvs(size=int(sample_size*0.9), type='float') + np.round(np.random.uniform(-20, 0, size=int(sample_size*0.1)), 2).tolist()
random.shuffle(powerlaw_rand_a_flt)
random.shuffle(powerlaw_rand_b_flt)
# powerlaw_rand_a = np.random.normal(10, 40, sample_size)
# powerlaw_rand_b = np.random.normal(10, 40, sample_size)

arithmetic_text = []
for i in tqdm(range(sample_size)):

    # i = np.random.randint(0, sample_size)
    sample_generator_a = np.random.choice([f"{powerlaw_rand_a_int[i]}*{beta_rand_a[i]}", f"{powerlaw_rand_a_flt[i]}*{beta_rand_a[i]}"])
    
    # j = np.random.randint(0, sample_size)
    sample_generator_b = np.random.choice([f"{np.abs(powerlaw_rand_b_int[i])}*{beta_rand_b[i]}", f"{np.abs(powerlaw_rand_b_flt[i])}*{beta_rand_b[i]}"])
    
    sample = sample_generator_a+np.random.choice(["+", "-"])+sample_generator_b
    # res = eval(sample) + np.random.randint(-10, 10) + np.random.rand()  # noise arithmetic
    # res = np.random.randint(-199, 199) + np.random.rand()  # wrong arithmetic
    res = eval(sample)
    arithmetic_text.append(sample + '=' + str(np.around(res, 2)))
arithmetic_text[:5]
```

```python
chars = ['<BOS>', '<EOS>', '<PAD>', '<MASK>', '+', '-', '*', '=', '.'] + [str(i) for i in range(0, 301)] + ['0'+str(i) for i in range(1, 10)]
vocab_size = len(chars)
print("all the unique characters:", ' | '.join(chars))
print(f"vocab size: {vocab_size:,}")
# make a vocab file
# with open('./data/arithmetic/vocab_arithmetic.txt', 'w', encoding='utf-8') as fp:
#     fp.write('\n'.join(arithmetic_text))
```

```python
# create a mapping from characters to integers
stoi = { ch:i for i,ch in enumerate(chars) }
stoi['00'] = stoi['0']
itos = { i:ch for i,ch in enumerate(chars)}
def encode(eq):
    seperation = [token for token in re.split(r'(\-|\+|\*|\=|\.)', eq) if token != '']
    length = len(seperation)
    if length == 21:
        return [stoi['<BOS>']] + [stoi[c] for c in seperation] + [stoi['<EOS>']] # encoder: take a string, output a list of integers
    else:
        return [stoi['<BOS>']] + [stoi[c] for c in seperation] + [stoi['<EOS>']] + [stoi['<PAD>'] for _ in range(21-length)]

def encode_embedding(eq):
    seperation = [token for token in re.split(r'(\-|\+|\*|\=|\.)', eq) if token != '']
    return [stoi['<BOS>']] + [stoi[c] for c in seperation] # encoder: take a string, output a list of integers

def decode(l):
    return ''.join([itos[i] for i in l]) # decoder: take a list of integers, output a string
```

```python
arithmetic_token = []
for eq in arithmetic_text:
    arithmetic_token.append(encode(eq))

train_arithmetic_token = arithmetic_token[:int(len(arithmetic_token)*0.9)]
valid_arithmetic_token = arithmetic_token[int(len(arithmetic_token)*0.9):]
```

```python
from model import GPTConfig, GPT
from torch.nn import functional as F

device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
# load data in
train_ids = np.load("./data/train_eco_pl_full.npy")
val_ids = np.load("./data/val_eco_pl_full.npy")
print("trainset size is", train_ids.shape, "\nvalset size is", val_ids.shape)

def get_batch(dataset, batch_size=64):
    max_n = len(dataset)
    idx = np.random.randint(max_n, size=batch_size)
    x = torch.tensor(dataset[idx, :-1])
    y = torch.tensor(dataset[idx, 1:])  # x and y are overlapped, offset by 1
    return x, y

test_batch_x, test_batch_y = get_batch(train_ids, batch_size=2048)
test_batch_x.shape, test_batch_y.shape  # block size or context size is 15
```

```python
model_args = dict(n_layer=8, n_head=8, n_embd=280, block_size=val_ids.shape[1]-1,
                  bias=False, vocab_size=vocab_size, dropout=0.2)
config = GPTConfig(**model_args)
model = GPT(config)
model.to(device)
```

```python
optimizer = torch.optim.AdamW(model.parameters(), lr=2e-3)
model.train()

# val_batch_x, val_batch_y = get_batch(val_ids, batch_size=512)
# val_batch_x, val_batch_y = val_batch_x.to(device), val_batch_y.type(torch.LongTensor).to(device)

p_bar = tqdm(range(1000))
for ite in p_bar:
    train_batch_x, train_batch_y = get_batch(train_ids, batch_size=1024)
    train_batch_x, train_batch_y = train_batch_x.to(device), train_batch_y.type(torch.LongTensor).to(device)
    logit, loss, _ = model(train_batch_x, train_batch_y)
    # _, val_loss, _ = model(val_batch_x, val_batch_y)
    optimizer.zero_grad(set_to_none=True)
    loss.backward()
    optimizer.step()

    p_bar.set_postfix({'val_loss': loss})
    
torch.save(model, 'saved_model/arithmetic-GPT-eco-uni.pt')
```

```python
c13k_fp = "./data/choices13k/c13k_selections.csv"
c13k = pd.read_csv(c13k_fp)
# with pd.option_context('display.max_columns', None):
#     display(c13k)

c13k_problems = pd.read_json("./data/choices13k/c13k_problems.json", orient='index')
# join the gamble payout probability information against the entries in c13k 
c13k_w_gambles = c13k.join(c13k_problems, how="left")

# with pd.option_context('display.max_columns', None):
#     display(c13k_w_gambles)

cal_a = []
for i in range(len(c13k_w_gambles)):
    element_a_H = np.random.choice([str(c13k_w_gambles['Ha'].values[i])+'*'+str(c13k_w_gambles['pHa'].values[i]), 
                                   str(c13k_w_gambles['Ha'].values[i])+'*'+str(c13k_w_gambles['pHa'].values[i])])
    element_a_L = str(c13k_w_gambles['La'].values[i])+'*'+str(np.round(1-c13k_w_gambles['pHa'].values[i], 2))
    if element_a_L[0] == "-":
        element_a = element_a_H+element_a_L+'='
    else:
        element_a = element_a_H+'+'+element_a_L+'='
    cal_a.append(element_a)
c13k_w_gambles['cal_a'] = cal_a

cal_b = []
for i in range(len(c13k_w_gambles)):
    element_b_H = np.random.choice([str(c13k_w_gambles['Hb'].values[i])+'*'+str(c13k_w_gambles['pHb'].values[i]), 
                                   str(c13k_w_gambles['Hb'].values[i])+'*'+str(c13k_w_gambles['pHb'].values[i])])
    element_b_L = str(c13k_w_gambles['Lb'].values[i])+'*'+str(np.round(1-c13k_w_gambles['pHb'].values[i], 2))
    if element_b_L[0] == "-":
        element_b = element_b_H+element_b_L+'='
    else:
        element_b = element_b_H+'+'+element_b_L+'='
    cal_b.append(element_b)
c13k_w_gambles['cal_b'] = cal_b

cal_ab = []
for i in range(len(c13k_w_gambles)):
    element_a_H = str(c13k_w_gambles['Ha'].values[i])+'*'+str(c13k_w_gambles['pHa'].values[i])
    element_a_L = str(c13k_w_gambles['La'].values[i])+'*'+str(np.round(1-c13k_w_gambles['pHa'].values[i], 2))
    element_b_H = str(c13k_w_gambles['Hb'].values[i])+'*'+str(c13k_w_gambles['pHb'].values[i])
    element_b_L = str(c13k_w_gambles['Lb'].values[i])+'*'+str(np.round(1-c13k_w_gambles['pHb'].values[i], 2))
    if element_a_L[0] == "-":
        element_a = element_a_H+element_a_L
    else:
        element_a = element_a_H+'+'+element_a_L
        
    if element_b_L[0] == "-":
        element_b = element_b_H+'+'+element_b_L[1:]+'='
    else:
        element_b = element_b_H+'-'+element_b_L+'='
    element_ab = element_a+'-'+element_b
    cal_ab.append(element_ab)
c13k_w_gambles['cal_ab'] = cal_ab

c13k_w_gambles.loc[0:5]
```

```python
dataset = c13k_w_gambles[c13k_w_gambles['Block']==1].reset_index()
embedding_a = []
embedding_b = []
embedding_ab = []
diff_ab = []
for i in tqdm(range(len(dataset))):
    a_prompt = torch.tensor(encode_embedding(dataset['cal_a'].values[i]), device=device).reshape(1, -1)
    # a_embedding = np.mean(model.embedding(a_prompt).cpu()[0].numpy(), axis=0)
    a_embedding = model.embedding(a_prompt).cpu()[0].numpy()[-1]
    b_prompt = torch.tensor(encode_embedding(dataset['cal_b'].values[i]), device=device).reshape(1, -1)
    # b_embedding = np.mean(model.embedding(b_prompt).cpu()[0].numpy(), axis=0)
    b_embedding = model.embedding(b_prompt).cpu()[0].numpy()[-1]
    ab_prompt = torch.tensor(encode_embedding(dataset['cal_ab'].values[i]), device=device).reshape(1, -1)
    # ab_embedding = np.mean(model.embedding(ab_prompt).cpu()[0].numpy(), axis=0)
    ab_embedding = (model.embedding(ab_prompt).cpu()[0].numpy()[-1] + model.embedding(ab_prompt[-15:]).cpu()[0].numpy()[-1])/2
    embedding_a.append(a_embedding.tolist())
    embedding_b.append(b_embedding.tolist())
    embedding_ab.append(ab_embedding.tolist())
    diff_ab.append((b_embedding-a_embedding).tolist())

dataset['embedding_a'] = embedding_a
dataset['embedding_b'] = embedding_b
dataset['embedding_ab'] = embedding_ab
dataset['diff_ab'] = diff_ab
dataset
```
