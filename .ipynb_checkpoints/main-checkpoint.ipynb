## data preparation

```python
class power_law:
    def __init__(self, a, b, g):
        self.a = a
        self.b = b
        self.g = g
        
    def rvs(self, size=1, type='int'):
        """Power-law gen for pdf(x)\propto x^{g-1} for a<=x<=b"""
        r = np.random.random(size=size)
        ag, bg = self.a**self.g, self.b**self.g
        if type=='int':
            return np.int32((ag + (bg - ag)*r)**(1./self.g)).tolist()
        else:
            return np.round((ag + (bg - ag)*r)**(1./self.g), 2).tolist()

    def pdf(self, x):
        ag, bg = self.a**self.g, self.b**self.g
        return self.g * x**(self.g-1) / (bg - ag)

power = power_law(1, 131, 0.01)
sample = power.rvs(size=10000, type='float')
# sample
# sample = np.abs(np.random.normal(loc=10.0, scale=50, size=1000))
_, _, _ = plt.hist(sample, bins=100)
# plt.plot(np.arange(1, 118), power.pdf(np.arange(1, 118)))
```

```python
# generate training dataset
sample_size = 1000000
# probabilities from beta distribution
beta_rand_a = np.round(np.random.beta(0.5, 0.5, size=sample_size), 2)
beta_rand_b = np.round(np.random.beta(0.5, 0.5, size=sample_size), 2)
# values from power law distribution
power = power_law(1, 131, 0.01)
powerlaw_rand_a_int = power.rvs(size=int(sample_size*0.9), type='int') + [int(n) for n in np.random.uniform(-20, 0, size=int(sample_size*0.1))]
powerlaw_rand_b_int = power.rvs(size=int(sample_size*0.9), type='int') + [int(n) for n in np.random.uniform(-20, 0, size=int(sample_size*0.1))]
random.shuffle(powerlaw_rand_a_int)
random.shuffle(powerlaw_rand_b_int)
powerlaw_rand_a_flt = power.rvs(size=int(sample_size*0.9), type='float') + np.round(np.random.uniform(-20, 0, size=int(sample_size*0.1)), 2).tolist()
powerlaw_rand_b_flt = power.rvs(size=int(sample_size*0.9), type='float') + np.round(np.random.uniform(-20, 0, size=int(sample_size*0.1)), 2).tolist()
random.shuffle(powerlaw_rand_a_flt)
random.shuffle(powerlaw_rand_b_flt)
# powerlaw_rand_a = np.random.normal(10, 40, sample_size)
# powerlaw_rand_b = np.random.normal(10, 40, sample_size)

arithmetic_text = []
for i in tqdm(range(sample_size)):

    # i = np.random.randint(0, sample_size)
    sample_generator_a = np.random.choice([f"{powerlaw_rand_a_int[i]}*{beta_rand_a[i]}", f"{powerlaw_rand_a_flt[i]}*{beta_rand_a[i]}"])
    
    # j = np.random.randint(0, sample_size)
    sample_generator_b = np.random.choice([f"{np.abs(powerlaw_rand_b_int[i])}*{beta_rand_b[i]}", f"{np.abs(powerlaw_rand_b_flt[i])}*{beta_rand_b[i]}"])
    
    sample = sample_generator_a+np.random.choice(["+", "-"])+sample_generator_b
    # res = eval(sample) + np.random.randint(-10, 10) + np.random.rand()  # noise arithmetic
    # res = np.random.randint(-199, 199) + np.random.rand()  # wrong arithmetic
    res = eval(sample)
    # add <MASK>

    random_num = np.random.rand()
    if random_num >0.8:
        id_list = []
        for n, lt in enumerate(sample):
            if lt in ["*", "+", "-"]:
                id_list.append(n)
        amb_sample = sample[:id_list[0]+1] + "<AMB>" + sample[id_list[1]:id_list[2]+1] + "<AMB>"
    else:
        amb_sample = sample
            
    arithmetic_text.append(amb_sample + '=' + str(np.around(res, 2)))
arithmetic_text[:5]
```

```python
chars = ['<BOS>', '<EOS>', '<PAD>', '<AMB>', '+', '-', '*', '=', '.'] + [str(i) for i in range(0, 301)] + ['0'+str(i) for i in range(1, 10)]
vocab_size = len(chars)
print("all the unique characters:", ' | '.join(chars))
print(f"vocab size: {vocab_size:,}")
# make a vocab file
# with open('./data/arithmetic/vocab_arithmetic.txt', 'w', encoding='utf-8') as fp:
#     fp.write('\n'.join(arithmetic_text))
```

```python
# create a mapping from characters to integers
stoi = { ch:i for i,ch in enumerate(chars) }
stoi['00'] = stoi['0']
itos = { i:ch for i,ch in enumerate(chars)}
def encode(eq):
    seperation = [token for token in re.split(r'(\-|\+|\*|\=|\.)', eq) if token != '']
    length = len(seperation)
    if length == 21:
        return [stoi['<BOS>']] + [stoi[c] for c in seperation] + [stoi['<EOS>']] # encoder: take a string, output a list of integers
    else:
        return [stoi['<BOS>']] + [stoi[c] for c in seperation] + [stoi['<EOS>']] + [stoi['<PAD>'] for _ in range(21-length)]

def encode_embedding(eq):
    seperation = [token for token in re.split(r'(\-|\+|\*|\=|\.)', eq) if token != '']
    return [stoi['<BOS>']] + [stoi[c] for c in seperation] # encoder: take a string, output a list of integers

def decode(l):
    return ''.join([itos[i] for i in l]) # decoder: take a list of integers, output a string
```

```python
arithmetic_token = []
for eq in arithmetic_text:
    arithmetic_token.append(encode(eq))

train_arithmetic_token = arithmetic_token[:int(len(arithmetic_token)*0.9)]
valid_arithmetic_token = arithmetic_token[int(len(arithmetic_token)*0.9):]
```

```python
optimizer = torch.optim.AdamW(model.parameters(), lr=2e-3)
model.train()

# val_batch_x, val_batch_y = get_batch(val_ids, batch_size=512)
# val_batch_x, val_batch_y = val_batch_x.to(device), val_batch_y.type(torch.LongTensor).to(device)

p_bar = tqdm(range(10000))
for ite in p_bar:
    train_batch_x, train_batch_y = get_batch(train_ids, batch_size=1024)
    train_batch_x, train_batch_y = train_batch_x.to(device), train_batch_y.type(torch.LongTensor).to(device)
    logit, loss, _ = model(train_batch_x, train_batch_y)
    # _, val_loss, _ = model(val_batch_x, val_batch_y)
    optimizer.zero_grad(set_to_none=True)
    loss.backward()
    optimizer.step()

    p_bar.set_postfix({'val_loss': loss})
    
torch.save(model, 'saved_model/arithmetic-GPT-eco-uni.pt')
```

```python
def reg_m(y, x, regularization=False):
    ones = np.ones(len(x[0]))
    X = sm.add_constant(np.column_stack((x[0], ones)))
    for ele in x[1:]:
        X = sm.add_constant(np.column_stack((ele, X)))
    if not regularization:
        results = sm.OLS(y, X).fit()
    else:
        results = sm.OLS(y, X).fit_regularized(method='sqrt_lasso', alpha=0.2)
    return results

dataset = pd.concat([IT_choices[:4578], IT_choices[4640:]])  #4578
print(reg_m(dataset['bRate'].values, np.c_[
            np.array(dataset['embedding'].values.tolist()), 
            ].T).summary())
```

```python
# dataframe concat
nov_choices = pd.read_csv('data/intertemporal_choices/nov2020_covid_problems.csv', index_col=0)
june_choices = pd.read_csv('data/intertemporal_choices/june2020_covid_problems.csv', index_col=0)
march_choices = pd.read_csv('data/intertemporal_choices/march2020_covid_problems.csv', index_col=0)
april_choices = pd.read_csv('data/intertemporal_choices/april2020_covid_problems.csv', index_col=0)
covid_choices = pd.concat([march_choices, april_choices, june_choices, nov_choices], axis=0)
covid_choices[:10]

cal = []
ev = []
for i in range(len(covid_choices)):
    element_right = str(np.round(covid_choices['right_val'].values[i], 2))+'*'+str(np.round(0.85**covid_choices['right_time'].values[i], 2))
    element_left = str(np.round(covid_choices['left_val'].values[i], 2))+'*'+str(np.round(0.85**covid_choices['left_time'].values[i], 2))
    
    element = element_right+'-'+element_left+'='
    answer = eval(element[:-1])
    cal.append(element)
    ev.append(answer)
    
covid_choices['cal_ev'] = cal
covid_choices['ev'] = ev
```

```python
covid_choices_subset = covid_choices[covid_choices["right_val"]<=300]
covid_choices_subset = covid_choices_subset[covid_choices_subset["left_val"]<=300]
len(covid_choices_subset)
```
